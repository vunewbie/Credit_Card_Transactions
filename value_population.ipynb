```python
import matplotlib.pyplot as plt

# Data
cities = ['Merritt Island', 'Tucson', 'Hollywood', 'San Francisco', 'Oakland', 
          'San Jose', 'Memphis', 'Philadelphia', 'Indianapolis', 'Hollywood']
population = [42389, 843746, 35716, 776733, 399484, 894943, 650100, 1517550, 1225000, 1225000]
transaction_value = [3.8, 3.7, 3.6, 3.4, 3.3, 2.7, 2.6, 2.5, 2.4, 2.2]

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(population, transaction_value, color='b')

# Add city name to each point
for i, city in enumerate(cities):
    plt.text(population[i] + 50000, transaction_value[i] + 0.05, city, fontsize=9)

# Set title and labels
plt.title('Mi quan h gia Dn s v Tng gi tr giao dch ca cc Thnh ph', fontsize=14)
plt.xlabel('Dn s (ngi)', fontsize=12)
plt.ylabel('Tng gi tr giao dch (T ng)', fontsize=12)

# Show plot
plt.grid(True)
plt.show()
```
